```python
from sklearn.preprocessing import OrdinalEncoder
```

#ENCODING

```python
review_order=["Poor","Average","Good"]
education_order=["School","UG","PG"]
encoder=OrdinalEncoder(categories=[review_order,education_order])
```

#TRANSFORMIN AFTER ENCODING

```python
df[["review","education"]]=encoder.fit_transform(df[["review","education"]])
```
